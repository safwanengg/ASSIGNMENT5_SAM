```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = pd.read_csv("data/city_data.csv")
ride_data_to_load =pd.read_csv("data/ride_data.csv")
mcr = pd.merge(city_data_to_load, ride_data_to_load, on="city", how="inner")
mcr.groupby('city')
mcr.head()
# Read the City and Ride Data

# Combine the data into a single dataset

# Display the data table for preview
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types

# Build the scatter plots for each city types

# Incorporate the other graph properties

# Create a legend

# Incorporate a text label regarding circle size

# Save Figure
U = mcr.loc[mcr["type"]=="Urban"]
S = mcr.loc[mcr["type"]=="Suburban"]
R = mcr.loc[mcr["type"]=="Rural"]
U = U.groupby('city')
U_mean = pd.DataFrame(U["fare"].mean())
U_mean = U_mean.reset_index()
U_fare = U_mean['fare']
U_Rides = pd.DataFrame(U.count())
U_Rides = U_Rides.reset_index()
U_number_of_rides = U_Rides['driver_count']
plt.scatter(U_number_of_rides, U_fare, marker="o", facecolors="darkorange", edgecolors="black", s=U_number_of_rides*10, alpha=1)
S = S.groupby('city')
S_mean = pd.DataFrame(S["fare"].mean())
S_mean =S_mean.reset_index()
S_fare = S_mean['fare']
S_Rides = pd.DataFrame(S.count())
S_Rides = S_Rides.reset_index()
S_number_of_rides = S_Rides['driver_count']
plt.scatter(S_number_of_rides, S_fare, marker="o", facecolors="lightblue", edgecolors="black", s=S_number_of_rides*10, alpha=1)
R = R.groupby('city')
R_mean = pd.DataFrame(R["fare"].mean())
R_mean =R_mean.reset_index()
R_fare = R_mean['fare']
R_Rides = pd.DataFrame(R.count())
R_Rides = R_Rides.reset_index()
R_number_of_rides = R_Rides['driver_count']
plt.scatter(R_number_of_rides, R_fare, marker="o", facecolors="yellow", edgecolors="black", s=R_number_of_rides*10, alpha=1)
plt.title("Pyber Ride Sharing Data (2016)")
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare (s)")
plt.legend(["Urban", "Suburban", "Rural"], loc="best")
plt.grid()
```

```python
# Show plot
plt.show()
```

```python
TotalFares_U = sum(U['fare'].sum())
TotalFares_S = sum(S['fare'].sum())
TotalFares_R = sum(R['fare'].sum())
T = [TotalFares_U, TotalFares_S, TotalFares_R]
T = sum(T)
TF_U = round((TotalFares_U/T)*100,1)
TF_S = round((TotalFares_S/T)*100,1)
TF_R = round((TotalFares_R/T)*100,1)
labels = ["Urban","Rural","Suburban"]

# The values of each section of the pie chart
sizes = [TF_U, TF_R,TF_S]

# The colors of each section of the pie chart
colors = ["lightcoral", "yellow", "skyblue", ]

# Tells matplotlib to seperate the "Python" section from the others
explode = (0.15, 0, 0)
plt.axis("equal")
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=280)
plt.title("% of Total Fares by City Type")
```

## Total Rides by City Type

## Total Drivers by City Type
